```python
#Pandas to process and store the dataset
import pandas as pd
#SKlearn for the linear regression
from sklearn import linear_model
#Pyplot for plotting, along with some config to make plots bigger, and legible
import matplotlib.pyplot as plt
#Plot config begin
font = {'family' : 'DejaVu Sans',
        'weight' : '400',
        'size'   : 18}
plt.rc('font', **font)
plt.rcParams['figure.figsize'] = (20, 10)
plt.rcParams['figure.dpi'] = 100
#Plot config end
```

```python
#We import the dataset, from a text file
df = pd.read_fwf('brain_body.txt')
```

```python
#Check the dataframe
#head() shows the first 5 rows
df.head()
```

```python
#Setting Brain column to be X, and Body column to be Y
x_values = df[['Brain']]
y_values = df[['Body']]
```

```python
#Look at the values in plot
plt.scatter(x_values, y_values)
```

```python
#Create a object of the linear regression class
body_model = linear_model.LinearRegression()
#Calling the object, and fitting it on x and y values. This will create a st line and store it within
body_model.fit(x_values, y_values)
```

```python
plt.scatter(x_values, y_values)
#Here, in y, you can see that we are calling the predict function of the object, and passing in the x values
plt.plot(x_values, body_model.predict(x_values))

#Thats it! Linear regression is done!
#In the bottom right end, you can see that the model is kinda accurate. In the later corners, there are some data which is way off.
#You can ignore those for now :)
```

```python
#Umm, this shuld be pretty obvious. :eyes:

plt.scatter(x_values, y_values, color='red')
plt.plot(x_values, body_model.predict(x_values), color='yellow', linewidth=0.5, alpha=0.5)
plt.scatter(x_values, body_model.predict(x_values), color='blue')
```

```python
#Imma zoom into the small cluster of points on the bottom, cause well it will speak for itself
#This kinda seems like the model is doing something :P
#In the overall sense, the points far away from the line is too disturbing, making us think the model is not good
plt.ylim(-100,800)
plt.xlim(-100,600)
plt.scatter(x_values, y_values, color='red')
plt.plot(x_values, body_model.predict(x_values), color='yellow', linewidth=0.5, alpha=0.5)
plt.scatter(x_values, body_model.predict(x_values), color='blue')

```

```python
#Create a new table, with the initial values, along with predicted data, and error in the prediction.

df['Predicted'] = body_model.predict(x_values)
df['Error'] = df['Predicted'] - df['Body']
df
```

```python
#Predicting a new value using our model
#If you input the value directly to the model, it will give an error. Hence we need to do all these steps.
#The error is pretty self explanation.

import numpy as np
array = np.array(93.5)
body_model.predict(array.reshape(1, -1))
```
